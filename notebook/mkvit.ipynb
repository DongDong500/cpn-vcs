```python
import os
import numpy as np
from PIL import Image
from PIL import ImageOps
from matplotlib import pyplot as plt
```

```python
pth = '/home/dongik/datasets/cpn'
dst = '/home/dongik/datasets/cpn_vit'

for fname in os.listdir(os.path.join(pth, 'Images')):
    name = fname.split('.')[0]

    img = Image.open( os.path.join(pth, 'Images', fname) ).convert('RGB')
    mask = Image.open( os.path.join(pth, 'Masks', fname) ).convert('L')

    img = np.array(img, dtype=np.uint8)
    mask = np.array(mask, dtype=np.uint8)
    
    h, w = np.where(mask > 0)
    tl = (h.min(), w.min())
    rb = (h.max(), w.max())
    pnt = ( int((tl[0] + rb[0])/2), int((tl[1] + rb[1])/2) )
    img[pnt[0]-3:pnt[0]+3, pnt[1]-3:pnt[1]+3] = [0, 255, 0]

    flip_img = np.flip(img)
    flip_mask = np.flip(mask)

    flipud_img = np.flipud(img)
    flipud_mask = np.flipud(mask)

    fliplr_img = np.fliplr(img)
    fliplr_mask = np.fliplr(mask)

    Image.fromarray(img).save( os.path.join(dst, 'Images', fname) )
    Image.fromarray(mask).save( os.path.join(dst, 'Masks', fname) )

    Image.fromarray(flip_img).save( os.path.join(dst, 'Images', name + '_1.bmp') )
    Image.fromarray(flip_mask).save( os.path.join(dst, 'Masks', name + '_1.bmp') )

    Image.fromarray(flipud_img).save( os.path.join(dst, 'Images', name + '_2.bmp') )
    Image.fromarray(flipud_mask).save( os.path.join(dst, 'Masks', name + '_2.bmp') )

    Image.fromarray(fliplr_img).save( os.path.join(dst, 'Images', name + '_3.bmp') )
    Image.fromarray(fliplr_mask).save( os.path.join(dst, 'Masks', name + '_3.bmp') )

    # fig = plt.figure(figsize=(10, 20))
    # fig.add_subplot(4, 2, 1)
    # plt.imshow(img, )
    # fig.add_subplot(4, 2, 2)
    # plt.imshow(mask, cmap='gray')
    # fig.add_subplot(4, 2, 3)
    # plt.imshow(flip_img, )
    # fig.add_subplot(4, 2, 4)
    # plt.imshow(flip_mask, cmap='gray')
    # fig.add_subplot(4, 2, 5)
    # plt.imshow(flipud_img, )
    # fig.add_subplot(4, 2, 6)
    # plt.imshow(flipud_mask, cmap='gray')
    # fig.add_subplot(4, 2, 7)
    # plt.imshow(fliplr_img, )
    # fig.add_subplot(4, 2, 8)
    # plt.imshow(fliplr_mask, cmap='gray')
    # plt.show()
    # break
```
